```python
import numpy as np
np.random.seed(0)

import matplotlib.pyplot as plt

import sympy as sp
sp.init_printing(use_latex='mathjax') 
```

# Integration Example

$$I = \int_{0}^{\pi} x \sin x ~ dx$$

```python
# The indefinite integral

x = sp.symbols('x')
f = x * sp.sin(x)
f.integrate(x)
```

```python
# The definite integral

sp.integrate(f, (x, 0, sp.pi))
```

## Averages

```python
def equispacedAverage(npts):
    xi = np.linspace(0, np.pi, npts)
    fi = xi * np.sin(xi)
    return np.mean(fi)

def randomAverage(npts):
    xi = np.random.uniform(0, np.pi, npts)
    fi = xi * np.sin(xi)
    return np.mean(fi)
  
def vizPoints(npts, equi=True):

	if equi:		
		xi = np.linspace(0, np.pi, npts)
	else:
		xi = np.random.uniform(0, np.pi, npts)

	plt.axhline(y=0, c='gray', alpha=0.5)
	y = np.zeros(npts);
	plt.plot(xi, y, 'o')

	plt.axis('off')
	plt.xlim(-0.03, np.pi+0.03)
	plt.tight_layout()
```

```python
npts = 1000
print(equispacedAverage(npts), randomAverage(npts))
```

```python
vizPoints(10, True)
vizPoints(10, False)
```

# 2D Example

$$I = \int_{-1}^{1} \int_{-1}^{1} g(x_1, x_2) dx_1 dx_2,$$

where $$g(x) = \begin{cases} 1, \text{ if } x_1^2 + x_2^2 \leq 1 \\ 0, \text{ otherwise}.\end{cases}$$

Thus, the area of the circle is the average value $\bar{g}$ times the size of the domain, which in this case is $[-1, 1]^2 = 2 \times 2 = 4$.

Thus area of circle = $4 \bar{g}$. 

```python
def MCdarts(npts, isPlot=False): 
    """Function outputs average value of g,
   given npts # of darts to throw randomly

   input : #darts, visualize
   output: average value of g
       """
    x1 = np.random.uniform(-1,1,size=npts)
    x2 = np.random.uniform(-1,1,size=npts)
    cond = x1**2 + x2**2 <= 1
    if isPlot: # toggle to visualize darts 
        plt.plot(x1[cond],x2[cond],'o', mec='none', alpha=0.8) 
        plt.plot(x1[~cond],x2[~cond],'o', mec='none', alpha=0.8) 
        plt.xlim(-1,1); plt.ylim(-1,1) 
        plt.xlabel('$x_1$') 
        plt.ylabel('$x_2$') 
        plt.gca().set_aspect('equal')
        plt.tight_layout()

        plt.show()
                
    return np.sum(cond)/float(npts)

def equiDarts(npts, isPlot=False):
    """total number of points are npts = ngrid*ngrid"""
    
    ngrid = int(np.sqrt(npts))
    
    x = np.linspace(-1, 1, ngrid)
    x1, x2 = np.meshgrid(x,x)

    cond = x1**2 + x2**2 <= 1
    
    if isPlot:
        plt.plot(x1[cond],x2[cond],'o', mec='none', alpha=0.8)
        plt.plot(x1[~cond],x2[~cond],'o', mec='none', alpha=0.8)
        plt.xlim(-1,1)
        plt.ylim(-1,1)
        plt.xlabel('$x_1$')
        plt.ylabel('$x_2$')
        plt.gca().set_aspect('equal')

        plt.tight_layout()
        plt.show()
        
    npts = ngrid*ngrid
    
    y = np.sum(cond,axis=0)
    y = np.sum(y)
    
    return y/float(npts)
```

```python
npts = 5000
print('area {}'.format(4*MCdarts(npts,isPlot=True)))
```

```python
print ('area {}'.format(4*equiDarts(5000, isPlot=True)))
```
