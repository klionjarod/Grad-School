# Linear Congruential Generators

$$n_{i+1} = (a n_i) \mod m,$$

```python
def LinCongGen(a, m, n0, num):
    # returns "num" integers "n"  
    n = np.zeros(num, dtype=int)
    n[0] = n0
    
    for i in range(1,num):
        n[i] = a * n[i-1] % m # % = modulo operator
        
    return n
```

```python
print(LinCongGen(3, 14, 15, 50))
```

Useful to pick large $m$.

```python
a = 16807
m = 2147483647
x = LinCongGen(a, m, 12, 10000).astype(float)/m

n, bins, patches = plt.hist(x, 20, density=True)
```

# SciPy Statistics

Scipy statistics provides a handy interface to numerous random numbers

```python
from scipy.stats import lognorm

# standard distribution specfiers
sigma = 0.5
mu    = 1.0

# SciPy translation (from documentation)
s     = sigma
loc   = 0.
scale = np.exp(mu)
```

```python
# ppf = inverse of cdf
# finds relevant range over which PDF is nonzero
x = np.linspace(lognorm.ppf(0.001, s), 
                lognorm.ppf(0.9999, s), 100) 
plt.xlim(min(x), max(x))

# plot CDF and PDF
plt.plot(x, lognorm.pdf(x, s, loc, scale), lw=3, label='pdf')
plt.plot(x, lognorm.cdf(x, s, loc, scale), lw=3, label='cdf')
plt.legend()
```

```python
# sample from distribution
plt.plot(x, lognorm.pdf(x, s, loc, scale), lw=3, label='pdf')
plt.xlim(min(x), max(x))

r = lognorm.rvs(s, loc, scale, size=1000)
plt.hist(r, 50, density=True, histtype='stepfilled', alpha=0.4)
plt.legend(loc='upper left')
```

# Poisson and Binomial Distribution

Poisson distribution is useful for modeling occurrence of rare events.

It can be derived from the binomial distribution, in the limit of $N \rightarrow \infty$, $p \rightarrow 0$, but mean $Np = \lambda$.

```python
from scipy.stats import binom, poisson
```

```python
N, p = 100, 0.01
lam  = N*p
```

```python
x = np.arange(binom.ppf(0.0001, N, p), binom.ppf(0.9999, N, p))
plt.plot(x, binom.pmf(x, N, p), 'o-', ms=8, label='binomial')
plt.plot(x, poisson.pmf(x, lam), 's-', ms=8, label='poisson')
plt.legend()
```
