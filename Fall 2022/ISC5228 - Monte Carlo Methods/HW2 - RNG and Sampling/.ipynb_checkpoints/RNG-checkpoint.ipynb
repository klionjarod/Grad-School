# Problem 1: LCG Random Number Generator:

```python
def LCG(N, m, a, n0 = 1):
    """
    n0 = initial seed
    N = N random numbers to generate (int)
    m = modulus
    a = multiplier    
    """
    n = np.empty(N,)
    #assume n0 = 1 for all problems 
    n[0] = n0
    for i in range(1, N):
        n[i] = (a * n[i - 1]) % m
        
    return n
```

### Part I

```python
print("(a) ", LCG(10, 2**3, 2))
```

```python
print("(b) ", LCG(10, 2**3, 4))
```

(c) The sequence becomes 0 rapidly after few numbers generated

### Part III

```python
print(LCG(10, 2**4, 3))
print("For a = 3, m = 2^4, period = 4")
print(LCG(15, 2**5, 3))
print("For a = 3, m = 2^5, period = 8")
```

### Part IV

# Problem 2: Sampling 1D Distribution:  
### <center> $f(x) = 6x(1 - x), 0 \leq x \leq 1$ </center>

## (a) accept-reject method:

```python
xi = np.linspace(0, 1, 1000); fi = 6.* xi * (1. - xi)
plt.rcParams['figure.figsize'] = (13,8)
plt.rcParams['font.size'] = 16
plt.plot(xi, fi)
plt.xlabel("x")
plt.ylabel("f(x) = 6x(1-x)")
plt.show()
print('f_max = 1.5 at x = 0.5')
```

```python
def acceptReject(samples, isPlot = True):
    xmin = 0.
    xmax = 1.
    fmax = 1.5
    
    x = np.random.uniform(xmin, xmax, samples)
    u = np.random.uniform(0., fmax, samples)
    fx = 6.* x * (1. - x)
    
    y = x[u <= fx]
    
    if isPlot:
        xi = np.linspace(0, 1); fi = 6.* xi * (1. - xi)
        plt.plot(y, u[u < fx],'o', alpha=0.4) #all points below f(x)
        plt.plot(x[u > fx], u[u > fx],'o', alpha=0.4) #all pts above f(x)
        plt.plot(xi, fi,'k-') #f(x)
    
    return y
```

```python
#generate samples
x = acceptReject(10**5)
```

```python
plt.plot(xi, fi)
plt.hist(x, 20, density = True)
plt.xlabel("x")
plt.ylabel("f(x) = 6x(1-x)")
plt.show()
```
