```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
# Define the TISP class with usual functions
class TISPClassifier:
    def __init__(self, lambdas, iterations = 100):
        
        #initialize parameters
        self.iterations = iterations
        self.lambdas = lambdas
        
        self.train_errs = []
        self.test_errs = []
        self.likelihoods = []
        self.num_features = []
    
    
    def sigmoid(self, z):
        ''' 
        Sigmoid function

        Parameters:
        x : int, float or numpy array

        Returns:
        the sigmoid function applied to each element in the array
        '''
        return 1 / (1 + np.exp(-z))
    
    
    def log_likelihood(self, Z, y):
        '''
        Log likelihood of penalized logistic regression TISP for classification
        
        Parameters:
        Z : dot product of the regressor matrix, X, and weights, w (array)
        y : transformed labels from {-1,1} -> {0,1}
        '''
        return (1 / Z.shape[0])*sum(np.log(1 + np.exp(-y * Z)))
    
    
    def count_features(self):
        '''
        count the number of features to eventually find thresholds
        '''       
        return np.count_nonzero(self.weights)
    
    
    def fit(self, X, y, X_test, y_test):
        '''
        Fit data to the TISP model
        
        Parameters:
        X_ : 2D Regressor matrix (num_obs, num_feats)
        y_ : 1D array of transformed labels of the form {0, 1}
        
        Returns:
        training errors, test errors, number of features, and log likelihoods
        '''
        num_obs = X.shape[0]
        num_feats = X.shape[1]
        
        #Start the loop through the lambdas
        for i in range(len(self.lambdas)):
            #reset features for each lambda
            likelihoods = []
            #local error arrays
            local_train_errs = []
            local_test_errs = []
            
            # start with zero weights
            self.weights = np.zeros(num_feats)
            
            #loop for 100 iterations
            for j in range(self.iterations):
                z = np.dot(X, self.weights)
                y_pred = self.sigmoid(z)
                
                #calculate gradient
                gradient = np.dot(X.T, y - y_pred)
                
                #calculate weights
                self.weights = self.weights + (1 / num_obs) * gradient
                #and apply hard-thresholding penalty (0 if |x| <= lambda)
                self.weights[np.absolute(self.weights) <= self.lambdas[i]] = 0
                
                #now calculate log-likelihood and append array
                likelihoods.append(self.log_likelihood(z, y))
                
                #Calculate train and test errors and append respective arrays
                local_train_errs.append(1 - self.score(X, y))
                local_test_errs.append(1 - self.score(X_test, y_test))
            
            #Capture the iteration loop's values
            self.likelihoods.append(likelihoods)
            
            self.train_errs.append(local_train_errs)
            self.test_errs.append(local_test_errs)
            
            self.num_features.append(self.count_features())
    
    
    def predict_proba(self, X):

        z = np.dot(X, self.weights)
        probabilities = self.sigmoid(z)

        return probabilities
    
    
    def predict(self, X, threshold=0.5):
        # Thresholding probability to predict binary values
        binary_predictions = np.array(list(map(lambda x: 1 if x > threshold else 0, self.predict_proba(X))))
        
        return binary_predictions
    
    
    def score(self, X, y_true):
        #classification accuracy
        y_pred = self.predict(X)
        acc = accuracy_score(y_true, y_pred)
        
        return acc
```

# Problem 4.a: Gisette dataset

```python
gis_train = pd.read_csv("../datasets/Gisette/gisette_train.data", sep = ' ', header=None).dropna(axis=1)
gis_train_labels = np.where(np.ravel(pd.read_csv("../datasets/Gisette/gisette_train.labels", sep = ' ', header=None).values) == -1, 0, 1)
gis_test = pd.read_csv("../datasets/Gisette/gisette_valid.data", sep = ' ', header=None).dropna(axis=1)
gis_test_labels = np.where(np.ravel(pd.read_csv("../datasets/Gisette/gisette_valid.labels", sep = ' ', header=None).values) == -1, 0, 1)
```

```python
#Normalize training data in gisette to have mean 0 and standard deviation 1
sc_gis = StandardScaler()
sc_gis.fit(gis_train)
gis_train_norm = sc_gis.transform(gis_train)
#apply the same transformation to testing data
gis_test_norm = sc_gis.transform(gis_test)
```

```python
#Initialize and fit TISP model for gisette dataset
'''
(# of features, lambda) = (502, 0.03854), (300, 0.0529),
                          (101, 0.0877), (30, 0.133), (10, 0.19)
'''
lambdas_gis = [0.03854, 0.0529, 0.0877, 0.133, 0.19]
gis_model = TISPClassifier(lambdas_gis)
gis_model.fit(gis_train_norm, gis_train_labels, gis_test_norm, gis_test_labels)
```

```python
#Plot the train misclassification error vs. iteration number for 100 features
plt.rcParams['figure.figsize'] = [15,8]
plt.plot([i+1 for i in range(len(gis_model.train_errs[2]))], gis_model.train_errs[2], label="100 Features")

plt.xlabel("Number of Iterations", fontsize = 14)
plt.ylabel("Training Misclassification Error", fontsize = 14)
plt.title("Number of Iterations vs Training Misclassification Error:\nGisette dataset", fontsize=20)
plt.legend(fontsize=14)

plt.savefig("Gisette 100 Features")
plt.show()
```

```python
#Plot final train and test misclassification error vs. # of selected features
plt.rcParams['figure.figsize'] = [15,8]
plt.plot(gis_model.num_features, [item[-1] for item in gis_model.train_errs], label="Train Data")
plt.plot(gis_model.num_features, [item[-1] for item in gis_model.test_errs], label="Test Data")

plt.xlabel("Number of Features", fontsize = 14)
plt.ylabel("Misclassification Error", fontsize = 14)
plt.title("Misclassification Error vs Number of Features\nGisette Dataset", fontsize = 20)
plt.legend(fontsize = 14)

plt.savefig("Gisette Misclassification Error")
plt.show()
```

```python
gis_info = {"Lambdas": lambdas_gis,
            "Train Error": [item[-1] for item in gis_model.train_errs],
            "Test Error": [item[-1] for item in gis_model.test_errs],
            "Number of Features": gis_model.num_features
            }
gis_df = pd.DataFrame(gis_info)
gis_df.sort_values(['Lambdas'], ascending=False).to_csv('Gisette Table.csv', index=False)
```
