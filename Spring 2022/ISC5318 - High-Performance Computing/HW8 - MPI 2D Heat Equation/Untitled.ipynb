```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
heat_map = np.genfromtxt(".\\2D Heat Equation\\x64\\Debug\\H_values.txt")
```

```python
plt.contourf(heat_map, origin = 'upper', extent = [0, 3, 0, 4])
plt.colorbar(label = "H(x, y, 20)")
plt.xlabel("x")
plt.xticks((0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0))
plt.ylabel("y")
plt.yticks((0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0))
plt.title("2D Heat Equation")
plt.savefig("contour.png")
plt.show()
```
